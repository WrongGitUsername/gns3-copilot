```python
# clean topology nodes data
from typing import Dict, Any
import copy
from pprint import pprint
import json

def clean_nodes_ports(data: dict) -> dict:
    """
    专门清洗你这种 nodes 数据结构
    把每个节点的 ports 列表精简为只保留 name 和 short_name
    """
    for node in data.values():                     # 遍历 R-1, R-2, R-3, R-4
        if "ports" in node and isinstance(node["ports"], list):
            node["ports"] = [
                {"name": port["name"], "short_name": port["short_name"]}
                for port in node["ports"]
            ]
    return data

clean_data = clean_nodes_ports(copy.deepcopy(example_data))
pprint(clean_data)
```
